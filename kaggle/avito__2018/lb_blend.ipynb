```python
import pandas as pd 
import os 
import seaborn as sns
%matplotlib inline 
```

```python
lb_pred = os.listdir('../lb_blend/')
df = pd.DataFrame()
for i, pred in enumerate(lb_pred):
    df[pred[:-4]] = pd.read_csv(os.path.join('../lb_blend/', pred)).deal_probability
    
df.head()
```

```python
sns.plt.figure(figsize=(15,13))
sns.heatmap(df.corr(),annot=True,fmt=".2f")
```

```python
sub = pd.read_csv('../input/sample_submission.csv')
sub['deal_probability'] = df.drop('best_public_blend', 1).mean(1)
sub.to_csv("../sub/lb_2.csv",index=False)
```
